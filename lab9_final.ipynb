Pero... antes de empezar, ¬øQue es `Bag of Words`?...

`Bag of Words` es un modelo de conteo utilizado en Procesamiento de Lenguaje Natural (NLP) que tiene como objetivo generar una representaci√≥n vectorial (vector de caracter√≠sticas en nuestro cas) para cada documento a trav√©s del conteo de las palabras que contienen.

La siguiente figura muestra un ejemplo de `Bag of Words` en acci√≥n:

<p align="center">
  <img src="https://user.oc-static.com/upload/2020/10/23/16034397439042_surfin%20bird%20bow.png" width="500">
</p>

Como pueden ver, el modelo de `Bag of Words` no resulta tan complicado, ¬øpero c√≥mo lo aplicamos en python?.

Como podr√°n darse cuenta del ejemplo anterior, para facilitar el conteo ser√° necesario transformar cada uno de los documentos en vectores, donde cada una de las posiciones posee un car√°cter. Este proceso es conocido como **tokenizaci√≥n** y lo podemos realizar de la siguiente forma:

````python
import nltk
nltk.download('punkt')


docs = ['The teacher rocks like a good rock & roll',
             'the rock is the best actor in the world']


docs_tokenizados = [word_tokenize(doc)  for doc in docs]
docs_tokenizados
````

Podemos mejorar un poco m√°s el proceso de tokenizaci√≥n agregando

- Stemming:  Definimos Stemming como un algoritmo basado en reglas que transforma las palabras a una forma general. Un ejemplo de stemming, es el siguiente:
- Eliminaci√≥n de Stopwords: Eliminaci√≥n de palabras muy frecuentes que entorpecen la clasificaci√≥n (por ejemplo, el, la los, la, etc...)

<p align="center">
  <img src="https://devopedia.org/images/article/218/8583.1569386710.png" width="300">
</p>


````python
# Definimos algunas stopword que queremos que sean eliminadas
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = stopwords.words('spanish')

# Definimos un tokenizador con Stemming
class StemmerTokenizer:
    def __init__(self):
        self.ps = PorterStemmer()
    def __call__(self, doc):
        doc_tok = word_tokenize(doc)
        doc_tok = [t for t in doc_tok if t not in stop_words]
        return [self.ps.stem(t) for t in doc_tok]

# Inicializamos tokenizador
tokenizador = StemmerTokenizer()

# Creamos algunos documentos
docs = ['The teacher rocks like a good rock & roll',
        'the rock is the best actor in the world',
        'New York is a beautiful city']

# Obtenemos el token del primer documento
[tokenizador(doc) for doc in docs]
````

````python
# Comparaci√≥n con el caso anterior
docs_tokenizados = [word_tokenize(doc) for doc in docs]
docs_tokenizados
````

#### Al Estilo Scikit

Scikit implementa `bag of words` a trav√©s de la clase `CountVectorizer()` la cual contiene muchas opciones para mejorar la tokenizaci√≥n.

````python
bow = CountVectorizer(tokenizer= StemmerTokenizer())
df = bow.fit_transform(docs)

pd.DataFrame(df.toarray(), columns=bow.get_feature_names_out())
````

Una de las cosas m√°s interesantes que provee son el use de n-gramas, los cuales, en palabras simples, son conjuntos de n-palabras que se concatenan entre si y que se consideran como tokens separados.

Pensemos en `Nueva York`. Cuando se tokeniza Nueva York, se generan dos tokens independientes que a simple vista no tienen relaci√≥n: `Nueva` `York`.
Al usar n-gramas (en un rango min=1,max=2) , generamos tanto `Nueva` y `York` como tambi√©n `Nueva York` como un token independiente.

````python
bow = CountVectorizer(tokenizer= StemmerTokenizer(), ngram_range=(1,2))
df = bow.fit_transform(docs)

pd.DataFrame(df.toarray(), columns=bow.get_feature_names_out())
````

De los resultados, podemos ver que generamos vectores de conteo para cada una de las palabras que conforman el corpus.  Un punto extra que se agrega en esta obtenci√≥n de frecuencias son los bigramas, que b√°sicamente son el conjunto de palabras de tama√±o de aparecen juntas en el texto.

````python
#generar el mismo pipeline
bog = CountVectorizer(tokenizer= StemmerTokenizer(), ngram_range=(1,2) # Este punto es opcional y es para generar bigramas
                      )

atributos_de_interes = ['intelligence_score', 'strength_score', 'speed_score', 'durability_score', 'power_score', 'combat_score']

escalador = MinMaxScaler()

transformaciones = ColumnTransformer([
    ('bag_of_words', bog, 'history_text'),
    ('min_max_scaler', escalador, atributos_de_interes)
])

classifier = MultinomialNB()

pipeline = Pipeline([
    ('transformaciones', transformaciones),
    ('classifier', classifier)
])

X_train, X_test, y_train, y_test = train_test_split(df_comics.drop('alignment', axis=1), df_comics['alignment'], test_size=0.33, random_state=42)

pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)

report = classification_report(y_test, y_pred)

print("Classification Report (MultinomialNB):")
print(report)
````

**Respuesta:**

````python
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import SelectPercentile
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC


classifiers = [
    {
        'classifier': RandomForestClassifier(),
        'params': {'n_estimators': [100, 200]}
    },
    {
        'classifier': LogisticRegression(),
        'params': {'C': [0.1, 1.0]}
    },
    {
        'classifier': SVC(),
        'params': {'C': [1.0, 10.0]}
    }
]

count_vectorizer_params = {
    'ngram_range': [(1, 1), (1, 2), (1, 3)],
    'max_df': [0.9, 1.0],
    'min_df': [1, 2]
}

percentiles = [20, 40, 60, 80]

pipeline = Pipeline([
    ('count_vectorizer', CountVectorizer()),
    ('feature_selection', SelectPercentile()),
    ('classifier', None)
])


best_scores = []
best_classifiers = []
best_params = []

for classifier in classifiers:
    pipeline.set_params(classifier=classifier['classifier'])

    param_grid = {
        'count_vectorizer__{}'.format(key): value for key, value in count_vectorizer_params.items()
    }
    param_grid.update({'classifier__{}'.format(key): value for key, value in classifier['params'].items()})

    search = GridSearchCV(
        estimator=pipeline,
        param_grid=param_grid,
        scoring='accuracy',
        cv=2
    )

    search.fit(X_train['history_text'], y_train)

    best_scores.append(search.best_score_)
    best_classifiers.append(search.best_estimator_.named_steps['classifier'])
    best_params.append(search.best_params_)

best_classifier_index = np.argmax(best_scores)
best_classifier = best_classifiers[best_classifier_index]
best_params = best_params[best_classifier_index]

print("Mejor clasificador encontrado:", best_classifier)
print("Mejores par√°metros encontrados:", best_params)

````
